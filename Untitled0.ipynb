PREGUNTA 2

```python
import yfinance as yf
import pandas as pd

# Descargar datos de Tesla
tesla = yf.Ticker("TSLA")

# Obtener ingresos trimestrales
income_statement = tesla.quarterly_income_stmt

# Extraer solo el Total Revenue y convertirlo a DataFrame
tesla_revenue = income_statement.loc[['Total Revenue']].T

# Restablecer el índice y renombrar columnas
tesla_revenue.reset_index(inplace=True)
tesla_revenue.columns = ['Date', 'Total Revenue']

# Mostrar las últimas 5 filas
print(tesla_revenue.tail())
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar datos de Tesla
tesla_data = yf.download('TSLA')
tesla_data.reset_index(inplace=True)

# Definir la función make_graph
def make_graph(data, title):
    plt.figure(figsize=(10,4))
    plt.plot(data['Date'], data['Close'], label='Precio de cierre')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de cierre (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Llamar a la función con los datos y título
make_graph(tesla_data, 'Precio de Cierre de las Acciones de Tesla')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Descargar los datos de GME
gme_data = yf.download('GME')
gme_data.reset_index(inplace=True)

# Definir la función make_graph
def make_graph(data, title):
    plt.figure(figsize=(10,4))
    plt.plot(data['Date'], data['Close'], label='Precio de cierre', color='purple')
    plt.xlabel('Fecha')
    plt.ylabel('Precio de cierre (USD)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Llamar a la función con los datos de GameStop
make_graph(gme_data, 'Precio de Cierre de las Acciones de GameStop')
```
